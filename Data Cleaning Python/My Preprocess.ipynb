# Worldcup Cricket Data Pre-Processing

```python
import pandas as pd
import json
```

```python
#Loading the data set
with open ('F:/Datasets/YouTube Analysis/Dhaval Pate/2 - Cricket Analytics/t20_json_files/t20_wc_match_results.json') as f :
    data = json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.info()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv('F:/Datasets/YouTube Analysis/Dhaval Pate/2 - Cricket Analytics/My Attempt/Cleaned CSV Files/My_Clean_match_summary.csv', index = False)
```

<h2 style="color:blue">(2) Process Batting Summary</h2>

```python
with open('F:/Datasets/YouTube Analysis/Dhaval Pate/2 - Cricket Analytics/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
df_bating = pd.DataFrame(all_records)
df_bating.tail()
    
```

```python
df_bating['out/not_out']= df_bating.dismissal.apply(lambda x:'out' if len(x)>0 else 'not out')
df_bating
```

```python
df_bating.drop(columns = ['dismissal'], inplace = True)
df_bating
```

```python
df_bating['batsmanName']=df_bating['batsmanName'].apply(lambda x:x.replace('â€',''))
df_bating['batsmanName'] = df_bating['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_bating.head(10)
```

```python
df_bating.info()
```

```python
df_bating.shape
```

```python
df_bating['match_id'] = df_bating['match'].map(match_ids_dict)
df_bating.head()
```

```python
df_bating.shape
```

```python
df_bating.to_csv('F:/Datasets/YouTube Analysis/Dhaval Pate/2 - Cricket Analytics/My Attempt/Cleaned CSV Files/My_Clean_fact_battibg_summary.csv', index = False)
```
